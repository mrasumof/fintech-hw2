```python
import pandas as pd
```

```python
menu_df = pd.read_csv("menu_data.csv", header=0)
menu_df
```

```python
sales_df = pd.read_csv("sales_data.csv", header=0)
sales_df
```

```python
joinned = sales_df.merge(menu_df, left_on="Menu_Item", right_on="item")
joinned
```

```python
quantity_menu_item = joinned.groupby("Menu_Item")["Quantity"].sum()
quantity_menu_item_df = pd.DataFrame(quantity_menu_item).reset_index()
quantity_menu_item_df
```

```python
joinned["revenue"] = joinned["Quantity"] * joinned["price"]
joinned["cogs"] = joinned["Quantity"] * joinned["cost"]
joinned
```

```python
revenue_menu_item = joinned.groupby("Menu_Item")["revenue"].sum()
revenue_menu_item_df = pd.DataFrame(revenue_menu_item).reset_index()
revenue_menu_item_df
```

```python
cogs_menu_item = joinned.groupby("Menu_Item")["cogs"].sum()
cogs_menu_item_df = pd.DataFrame(cogs_menu_item).reset_index()
cogs_menu_item_df
```

```python
report_df = revenue_menu_item_df.merge(cogs_menu_item_df, left_on="Menu_Item", right_on="Menu_Item" )
report_df
```

```python
report_df2 = report_df.merge(quantity_menu_item_df, left_on="Menu_Item", right_on="Menu_Item" )
report_df2
```

```python
report_df2["profit"] = report_df2["revenue"] - report_df2["cogs"]
report_df2
```

```python
report_df2.columns = ["Menu_Item", "02-revenue", "03-cogs", "01-count", "04-profit"]
report_df2.set_index("Menu_Item", inplace=True)
report_df2
```

```python
report_df2 = report_df2[["01-count","02-revenue","03-cogs","04-profit"]]
report_df2
```

```python
report_dict = report_df2.T.to_dict()
report_dict
```

```python
f = open("PyRamen_report.txt","w")
f.write(str(report_dict))
f.close()
```
